```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/appliedcode/mthree-c422/main/Exercises/day-2/Simple%20Linear%20Regression/homeprices.csv', low_memory=False)
df
```

```python
plt.figure(figsize = (4,3))
plt.xlabel('area(ft^2)')
plt.ylabel('price($)')
plt.scatter(df['area'], df['price'], marker = '+')
plt.show()
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']], df[['price']])
```

```python
model.predict([[3300]])
```

```python
coef, bias = model.coef_, model.intercept_
coef, bias
```

```python
plt.figure(figsize = (4,3))
plt.xlabel('area(ft^2)')
plt.ylabel('price($)')
plt.scatter(df[['area']], df[['price']], marker = '+')
plt.plot(df[['area']], model.predict(df[['area']]), color = 'red')
plt.show()
```

```python
df_area = pd.read_csv('https://raw.githubusercontent.com/appliedcode/mthree-c422/main/Exercises/day-2/Simple%20Linear%20Regression/areas.csv')
df_area.sample(3)
```

```python
p = model.predict(df_area)
p
```

```python
df_area['prices'] = p
```
